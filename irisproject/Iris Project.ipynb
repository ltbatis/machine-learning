## Visualization

```python
sns.pairplot(iris, hue='species', height=2, aspect=1);
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.violinplot(x='class', y='sepal length', data=df)
plt.subplot(2,2,2)
sns.violinplot(x='class', y='sepal width', data=df)
plt.subplot(2,2,3)
sns.violinplot(x='class', y='petal length', data=df)
plt.subplot(2,2,4)
sns.violinplot(x='class', y='petal width', data=df)
```

```python
iris.boxplot(by='species', figsize=(12,8));
```
